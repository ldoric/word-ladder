# Puzzle catalog (`data/puzzles/catalog.json` / web game)

Regenerate the curated ladders from **`\*_strict_largest_island.txt`** (both endpoints and every step live in that list; graph = one-letter edits inside the vocabulary).

- **Easy:** BFS distance 3–5 edges (four puzzles per lang/length: 3,4,4,5).
- **Medium:** 6 or 7 edges (two at 6, two at 7).
- **Hard:** 8 edges (four puzzles).

Canonical generator: `scripts/generate_puzzle_catalog.py` (run from repo root).

```python
from pathlib import Path
import sys

REPO = Path.cwd().resolve()
if REPO.name == "notebooks":
    REPO = REPO.parent
if str(REPO) not in sys.path:
    sys.path.insert(0, str(REPO))

from scripts.generate_puzzle_catalog import build_catalog, DEFAULT_OUT

SEED = 42
catalog = build_catalog(SEED)
DEFAULT_OUT.write_text(
    __import__("json").dumps(catalog, ensure_ascii=False, indent=2) + "\n",
    encoding="utf-8",
)
print("Wrote", len(catalog["puzzles"]), "puzzles to", DEFAULT_OUT.relative_to(REPO))
```

```python
import pandas as pd

df = pd.DataFrame(catalog["puzzles"])
df["steps"] = df["solution"].apply(len)
display(
    df.groupby(["locale", "wordLength", "difficulty"])["id"]
    .count()
    .rename("n")
    .reset_index()
)
df.head(8)
```
